```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
fifa_best=pd.read_csv(r'C:\Users\sanjay\Downloads\Fifa1.csv')
fifa_best.head()
```

```python
fifa_best.shape
```

```python
sns.barplot(x="OVERALL",y="PLAYER",data=fifa_best.head(20),hue="OVERALL")
plt.title("Top 10 players In Fifa 22 Edition")
plt.show()
```

```python
data=fifa_best['POSITION'].value_counts()
values=fifa_best['POSITION'].value_counts().keys()
plt.figure(figsize=(10,10))
plt.pie(data,labels=values,autopct='%0.01f%%')
plt.title('Percentage Of Top Players In Fifa With Respect To Their Positions')
plt.show()
```

```python
data=fifa_best['TEAM'].value_counts()
values=fifa_best['TEAM'].value_counts().keys()
plt.figure(figsize=(15,15))
plt.pie(data,labels=values,autopct='%0.01f%%')
plt.title('Top players percentage in their resepective clubs')
plt.show()
```

```python
#To find the Players whose nationality is brazil
brazil_case=fifa_best[fifa_best["NATIONALITY"]=="Brazil"] 
brazil_case.head()
```

```python
#Finding the number of players who is either striker,left winger OR rightwinger
striker_left_right=fifa_best[(fifa_best["POSITION"] =="Striker") | (fifa_best["POSITION"] =="Left Winger") | (fifa_best["POSITION"]=="Right Winger")]
striker_left_right.head(10)
```

```python
#Finding the number of players who pace is greater than 90
pace_player=fifa_best[fifa_best["PACE"]>=90] 
pace_player.head(10)
pace=pace_player.sort_values(by="PACE",ascending=False)
```

```python
sns.barplot(x="PLAYER",y="PACE",data=pace.head(10),hue="PACE")
plt.title("Top 10 PACERS In Fifa 22 Edition")
plt.xticks(rotation=45)
plt.show()
```

```python
#Finding the number of players who passing is greater than 90
pass_player=fifa_best[fifa_best["PASSING"]>=85] 
pass_player.head(10)
passing=pass_player.sort_values(by="PASSING",ascending=False)
```

```python
sns.barplot(x="PLAYER",y="PASSING",data=passing.head(10),hue="PASSING")
plt.title("Top 10 Creaters In Fifa 22 Edition")
plt.xticks(rotation=60)
plt.show()
```

```python
#Finding the number of players who passing is greater than 90
defence_player=fifa_best[fifa_best["DEFENCE"]>=85] 
defence_player.head(10)
defence=defence_player.sort_values(by="DEFENCE",ascending=False)
```

```python
sns.barplot(x="PLAYER",y="DEFENCE",data=defence.head(10),hue="DEFENCE")
plt.title("Top 10 DEFENDERS In Fifa 22 Edition")
plt.xticks(rotation=45)
plt.show()
```
